```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bsort(u,v):
    n=len(u)
    for i in range(n-1):
        for j in range(n-i-1):
            if u[j]>u[j+1]:
                u[j], u[j+1]=u[j+1], u[j]
                v[j], v[j+1]=v[j+1], v[j]
    return u, v
```

```python
from sklearn.metrics import mean_squared_error as mse
index=[]
errors=[]
for i in range(70):
    Y_test=np.loadtxt('Test_N'+str(i)+'.csv')[0,:]
    pred=np.loadtxt('TestPredict_N'+str(i)+'.csv')[0,:]
    index.append(i)
    errors.append(mse(Y_test,pred))
errors, index=bsort(errors,index)
```

```python
for i in range(5):
    node=index[i]
    Y_test=np.loadtxt('Test_N'+str(node)+'.csv')[0,:]
    pred=np.loadtxt('TestPredict_N'+str(node)+'.csv')[0,:]
    print('error: ',errors[i])
    plt.plot(pred,label='predictions')
    plt.plot(Y_test,label='exact')
    plt.legend()
    plt.show()
```

```python
for i in range(5):
    node=index[-(i+1)]
    Y_test=np.loadtxt('Test_N'+str(node)+'.csv')[0,:]
    pred=np.loadtxt('TestPredict_N'+str(node)+'.csv')[0,:]
    print('error: ',errors[i])
    plt.plot(pred,label='predictions')
    plt.plot(Y_test,label='exact')
    plt.legend()
    plt.show()
```
